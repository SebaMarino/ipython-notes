```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as pyfits
import math as ma
%matplotlib inline
```

```python
def BBody(lam,T):
    #lam in cm
    h_planck=6.62607e-27 # cgs 
    k=1.381e-16          #erg/K
    c=3.0e10             # cm/s
    
    return 2.0*h_planck*c/(lam**3.0*(ma.exp(h_planck*c/(lam*k*T))-1.0))
```

```python
R_sun=6.955e+10 # cm
R_star=3.0*R_sun
T_star=6250.0
pc=3.085677e18 # cm

SED=np.zeros((17,3)) # dereddened sed
SED[:,0]=[0.36, 0.44, 0.55, 1.23, 1.65, 2.22, 3.77, 4.78, 11.85, 18.72, 70, 100, 160, 800, 1200, 1300, 3476]  
SED[:,1]=[1.8709094e-11, 3.6059398e-11, 3.1669204e-11, 8.8363268e-12, 5.5963674e-12, 3.6237149e-12, 1.6604315e-12, 9.3517083e-13, 1.9985633e-13, 1.3286577e-13, 7.055132e-14, 3.0392389e-14, 7.8856171e-15, 2.0928652e-17, 2.4804328e-18, 2.245599e-18, 1.2405385e-20]  
SED[:,2]=[5.5026747e-13, 3.2195891e-13, 3.4423048e-13, 2.3253491e-13, 1.2166016e-13, 6.4709194e-14, 6.731479e-14, 4.1873321e-14, 1.3626568e-14, 1.3650593e-14, 3.5145491e-15, 1.5307814e-15, 3.9864099e-16, 2.4915062e-18, 4.4293443e-20, 5.6611738e-20, 1.5836662e-21] 

#SED=np.loadtxt('sed_obs.dat')
c=2.99792458e18 # light speed in A/s
#SED[:,1]=SED[:,1]*c/((SED[:,0]*1.0e4)**2.0)

SED1=np.loadtxt('sed_teo0.dat') #  4.5e-10 M_sun in erg/s Hz cm**2 at 1 pc
SED1[:,1]=SED1[:,1]*c/((SED1[:,0]*1.0e4)**2.0) # in erg/s A cm**2 
SED1[:,1]=SED1[:,1]*1.0e4 # in erg/s um cm**2.0

SED2=np.loadtxt('sed_bbod.dat') #  4.5e-10 M_sun in erg/s Hz cm**2 at 1 pc
SED2[:,1]=SED2[:,1]*c/((SED1[:,0]*1.0e4)**2.0) # in erg/s A cm**2 
SED2[:,1]=SED2[:,1]*1.0e4 # in erg/s um cm**2.0

SED_star=np.zeros(len(SED1[:,1]))
for i in xrange(len(SED1[:,1])):
    wave=SED1[i,0]
    SED_star[i]=BBody(wave*1.0e-4,T_star)*ma.pi*(R_star/(140.0*pc))**2.0 # Fnu in cgs

SED_star=SED_star*1.0e-3*1.0e4*c/((SED1[:,0]*1.0e4)**2.0) # Flam in mks
```

```python
d=140.0 #pc

C=1.0e-3/(d**2.0) # erg/cm^2 to joule/m^2 and at 140 pc

plt.figure(figsize=(15,5))

plt.errorbar(SED[:,0],SED[:,1]*SED[:,0],yerr=(SED[:,2]*SED[:,0]),color='blue',fmt='o',label='Data')

plt.plot(SED1[:,0],SED1[:,1]*C*SED1[:,0],color='red', label='Model')
plt.plot(SED1[:,0],SED_star[:]*SED1[:,0],color='black', label='Black body')

#plt.plot(SED2[:,0],SED2[:,1]/SED2[:,0],'o',color='blue')

plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'Wavelength [$\mu m$]')
plt.ylabel('$\lambda F_{\lambda}$ [ $W$ $m^{-2}$]')
plt.ylim(1.0e-19,5.0e-11)
plt.xlim(0.3,5.0e3)
plt.legend(numpoints=1)
```
