# Disk Vortex model from Lyra 

This is an implementation of steady state dust distribution in disk vortices proposed in Lyra & Min-Kai 2013. 
They used the coordinate system $(x,y)$ accordint to Chand & Oishi (2010) in which they are centered at the vortex.

##### $$ \begin{align}  x &=& a \cos(\nu) \\  y&=& a\chi \sin(\nu) \\  \end{align} $$

#### $$ \rho(a,z) = \varepsilon \rho_{0} (S+1)^{3/2} \exp \left\{ - \frac{[a^{2} f^{2}(\chi)+z^{2}]}{2H^{2}}(S+1)  \right\}  $$

$$ \begin{align}
f^2(\chi) &=& 2\omega_{v}\chi - \xi^{-1}_{+} (2\omega_{v}^{2}+3) \\
 \xi_{\pm} &=& 1 \pm \chi^{-2} \\
 S &=& \frac{\mathrm{St}}{\delta} \\
 \mathrm{St} &=& \sqrt{\frac{\pi}{8}} \frac{a_{\bullet}}{H} \frac{\rho_{\bullet}}{\rho_{g}}
 \end{align} $$   

where a is the semi-minor axis, $\chi$ is the aspect ratio and $\omega_{v}$ is the ratio between the angular velocity of the vortex and the disk.

```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.colors import LogNorm
import math as ma
import astropy.io.fits as pyfits
```

```python
#------------------------- PARAMETERS AND SOME DEFINITIONS
"""
x     =    a*cos(nu)         elliptical coordinate (minor axis) 
y     =    a*chi*sin(nu)     elliptical coordinate  (mayor axis)
a     =                      vortex semi-minor axis
rho_0 =                      rho_gas,max
ep    =                      dust-to-gas ratio (global)
chi   =                      Vortex aspect ratio  (>1)
H     =   c_s/Omega          Sonic scale or gas scale height  
D     =   D=delta*c_s*H      Dust difussion coeficient 
S     =   St/delta           Dimensionless number where St is the stoke number and delta is the dimensionless difussion parameter
w_v   =   Omega_v/Omega      Dimensionless vortex frequency                   
a     =                      vortex semi-minor axis 
delta =   D=delta*c_s*H      Dimensionless diffusion parameter
"""
    
au=1.496e11                         # Astronomical unit in m
ep    =  0.01                       # global dust to gas ratio
rho_0 =  1.0
chi   =  4.0
delta= 10.0**-3                     # dimensionless difussion parameter
w_v   =  3.0/(2*(chi-1))            # Kida 1981 eq 9 Omega_V/Omega=2/3(chi-1)
#w_v   =  ma.sqrt(3.0/(chi**2.0-1)) # Goodman-Narayan-Goldreich solution
xi_p  = 1.0 + chi**(-2.0)
f_chi = ma.sqrt(2.0*w_v*chi - (2.0*w_v**2.0 + 3.0)/xi_p)
scale_h = 0.1                       # H/r = scale_h

r_0=5                               #au position of the vortex center               
```

```python
def rho_1(a,zi,ep,rho_0,S,chi,H,f_chi):
        
    return   ep*rho_0*(S+1)**1.5*ma.exp( - ( (a*f_chi)**2.0  + zi**2.0)*(S+1)/(2*H**2.0))
    
def Phi(xi,yi):

    # 1st quadrant
    if xi>0.0 and yi>=0.0:
        return ma.atan(yi/xi)
    
    if xi==0.0 and yi>=0.0: return ma.pi/2.0
    
    # 2nd quadrant
    if xi<0.0 and yi>=0.0:
        return ma.pi+ma.atan(yi/xi)

    # 3rd quadrant
    if xi<0.0 and yi<0.0:
        return -ma.pi+ma.atan(yi/xi)
    
    if xi==0.0 and yi<0.0: return -ma.pi/2.0
    
    # 4th quadrant
    if xi>0.0 and yi<0.0:
        return ma.atan(yi/xi)
```

```python
# ------------------------------ PLOTING VORTEX IN ELIPTICAL COORDINATES for S=0 (tracing the gas)

S=0.0

N=300
Nz=1
rho_el=np.zeros((N,N,Nz))
x=np.zeros(N)
y=np.zeros(N)
z=np.zeros(N)

dx, dy = 10.0/N, 20.0/N  # 
dz=0.1*dx   # Deltaz = 0.5 au

for i in range(N):
    for j in range(N):
        for k in range(Nz):
            
            x[i] = (i-(N-1)/2)*dx
            y[j] = (j-(N-1)/2)*dy
            z[k] = (k-(Nz-1)/2)*dz
            H=scale_h*(r_0+x[i])
            a = ma.sqrt(x[i]**2.0+(y[j]/chi)**2.0)
               
            rho_el[j,i,k] =  rho_1(a,z[k],ep,rho_0,S,chi,H,f_chi)
plt.figure(figsize=(7,3)) 
plt.pcolor(y,x,rho_el[:,:,(Nz-1)/2.0].T)  # z=0
plt.ylim(x[0],x[N-1])
plt.xlim(y[0],y[N-1])
plt.xlabel('y')
plt.ylabel('x')
plt.colorbar()
```

Changing this solution to cartesian coordiantes $(u,v)$ from the elliptical coordinates $(x,y)$.

### $$ \begin{align}  x & = & r-r_{0} \\   y & = & r\theta \\ u & = & r \cos(\theta) \\ v & = & r \sin(\theta)  \\  x&=&\sqrt{u^2+v^2}-r_{0} \\ y&=& r \tan^-1(v/u) \end{align} $$

```python
#---------------------------------------Evaluatin the gas density in cartesian coordinates

N=301                               # N odd. N x N x Nz is the dimension of the data cube
Nz=1  
S=0.0
rho_0=1.0e-10       # gas maximum density [g/cm^3]
ep=1.0
u=np.zeros(N)
v=np.zeros(N)
z=np.zeros(Nz)

du=dv = 3.0*r_0/N
dz=3.0*r_0*scale_h/Nz # 3 sigmas 

rho_gas=np.zeros((Nz,N,N))

for i in range(N):
    for j in range(N):
        
        u[i] = (i-(N-1)/2)*du
        v[j] = (j-(N-1)/2)*dv
        r=ma.sqrt(u[i]**2.0+v[j]**2.0)
        theta=Phi(u[i],v[j])
        xj=r-r_0
        yj=r*theta
        H=scale_h*r
        a = ma.sqrt(xj**2.0+(yj/chi)**2.0)
        a_max=2.0*H/(chi*w_v)   # artificialy increased a_max eq 68 in Lyra et al. 2013
        
        
        if  a<a_max:  # Inside the vortex
            for k in range(Nz):
                z[k] = (k-(Nz-1)/2)*dz
                rho_gas[k,j,i]=rho_1(a,z[k],ep,rho_0,S,chi,H,f_chi)        
        elif r>0.0:    # outside the vortex we fixed a to a_max
            for k in range(Nz):
                z[k] = (k-(Nz-1)/2)*dz
                rho_gas[k,j,i]=rho_1(a_max,z[k],ep,rho_0,S,chi,H,f_chi)
```

```python
#---------------PLOTING
Rho=rho_gas[(Nz-1)/2,:,:]

plt.figure(figsize=(7.5,6))
plt.title("z = %0.1f au" % z[(Nz-1)/2])
plt.pcolor(u,v,Rho,vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(u[0],u[N-1])
plt.ylim(v[0],v[N-1])
plt.xlabel('u [au]')
plt.ylabel('v [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),0.5,color="white",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',format='%1.1e',label=r'$\rho_{gas}$')
```

## Dust distribution

where $ a_{\bullet}$  is the dust size and $ \rho_{\bullet}$ is its internal density. We neglect the size restriction of the vortex in order to show the different distribution for different sizes.
