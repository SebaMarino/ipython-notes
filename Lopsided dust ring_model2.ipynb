# Implementation of a Lopsided dust ring

This is an implementation of the analysis made in Birnstiel, T. et al. 2013 of a lopsided dust ring. First we construct a axi-symetric dust gas distribution
$$ \bar{\rho}_g (r,z)= \frac{\bar{\Sigma}_{g}(r)}{\sqrt{2\pi}H}\exp\left[ -\frac{z^2}{2H^2}\right] $$
where $H=c_{s}/ \Omega$ is the scale hight which is typically $0.1 r$.

$$  \bar{\Sigma}_{g}(r)=\Sigma_{c} \left(\frac{r}{r_{c}} \right) ^{-\gamma} \times \exp\left( -\frac{1}{2(2-\gamma)} \left[ \left(\frac{r}{r_{c}} \right)^{2-\gamma} -1 \right] \right) $$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.colors import LogNorm
import math as ma
import astropy.io.fits as pyfits
```

```python
#----------------------------------------------------PARAMETERS

au=1.496e13           # [cm]

h_scale=0.1           # scale height H=h_scale*r

rc=120.0
sigmac=10.0 # g/cm^2
gamma=-5.15  # the more negative gamma is, the more sharp the ring is
Rmin=30.0


N=301
Nz=51
Rmax=300              # [au]  max x or y of the sampling
Zmax=3.0*Rmax*h_scale # [au]  max z of the zampling
dx=dy=Rmax*2.0/(N-1)
dz=Zmax*2.0/(Nz-1)
```

```python
#----------------------------------------------------FUNCTIONS

def Phi(xi,yi):

    # 1st quadrant
    if xi>0.0 and yi>=0.0:
        return ma.atan(yi/xi)
    
    if xi==0.0 and yi>=0.0: return ma.pi/2.0
    
    # 2nd quadrant
    if xi<0.0 and yi>=0.0:
        return ma.pi+ma.atan(yi/xi)

    # 3rd quadrant
    if xi<0.0 and yi<0.0:
        return -ma.pi+ma.atan(yi/xi)
    
    if xi==0.0 and yi<0.0: return -ma.pi/2.0
    
    # 4th quadrant
    if xi>0.0 and yi<0.0:
        return ma.atan(yi/xi)

#-----Axisymetric density    

def Sigma_p(r,rc,sigmac,gamma):
    
    return sigmac*(r/rc)**(-gamma)*ma.exp(-((r/rc)**(2.0-gamma)-1.0)/(2.0*(2.0-gamma)))

def rho_p(r,z,h_scale,rc,sigmac,gamma):
    
    H=r*h_scale
    return Sigma_p(r,rc,sigmac,gamma)*ma.exp(-z**2.0/(2.0*H**2.0))/(ma.sqrt(2.0*ma.pi)*H*au)
```

```python
#----------------------------------------------------MAIN

rhop_g=np.zeros((Nz,N,N))
X=np.linspace(-Rmax,Rmax,N)-dx/2.0 
Y=np.linspace(-Rmax,Rmax,N)-dy/2.0
Z=np.linspace(-Zmax,Zmax,Nz)-dz/2.0

#--------Axisymetric solution
for j in range(N):
    for i in range(N):
        x=(i-(N-1)/2.0)*dx
        y=(j-(N-1)/2.0)*dy
        r=ma.sqrt(x**2.0+y**2.0)
        if r>0.0:
            for k in range(Nz):
                z=(k-(Nz-1)/2.0)*dz
            
                rhop_g[k,j,i]=rho_p(r,z,h_scale,rc,sigmac,gamma)
```

```python
#-----------------------------PLOTTING
Rho=rhop_g[(Nz-1)/2,:,:]

plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
plt.title("z = %0.1f au" % 0.0)
plt.pcolor(X,Y,Rho,cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/cm$^3$]')

plt.subplot(2,2,2)
plt.title("z = %0.1f au, y = %0.1f au" % (0.0, 0.0))
plt.plot(X,Rho[(N-1)/2,:])
plt.xlim(X[0],X[N-1])
plt.xlabel("X")
plt.ylabel(r'$\rho_{g}$ [g/cm$^3$]')

Rho=rhop_g[:,(N-1)/2,:]

plt.subplot(2,1,2)
plt.title("y = %0.1f au" % 0.0)
plt.pcolor(X,Z,Rho,cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0],X[N-1])
plt.ylim(Z[0],Z[Nz-1])
plt.xlabel('X [au]')
plt.ylabel('Z [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/cm$^3$]')
```

## Non-axisymetric distribution

The non-axysimetric model is
$$  \rho_{g}(r,\phi)=\bar{\rho}_{g}(r)\left[1+A(r)\sin\left(\phi+\frac{\pi}{2}\right) \right]  $$

$$  A(r)= \frac{c-1}{c+1}\exp\left[-\frac{(r-R_{s})^2}{2H^2}\right] $$
where $c=\Sigma_{g,max}/\Sigma_{g,min}$ is the largest contrast of the gas surface density and $R_{s}$ is the position of the radial preassure bump.

```python
#----------------------------------------------------PARAMETERS

Rs=rc*1.3 # [au] Position of the radial preassure bump. Rs has to be slightly higher than rc
c=5.0                #largest contrast of the gas surface density
```

```python
#----------------------------------------------------FUNCTIONS
    
#-------Non-axisymetric density    
def A(r,c,Rs,h_scale):
    
    H=h_scale*r
    #return (c-1.0)*ma.exp(-(r-Rs)**2.0/(2.0*H**2.0))/(c+1.0)
    return (c-1.0)/(c+1.0)
def rho_gas(r,phi,rhop,c,Rs,h_scale):
    
    return rhop*(1.0+A(r,c,Rs,h_scale)*ma.sin(phi-ma.pi/2.0))
```

```python
#-------------Non-axisymetric solution

rho_g=np.zeros((Nz,N,N))

for j in range(N):
    for i in range(N):
        x=(i-(N-1)/2.0)*dx
        y=(j-(N-1)/2.0)*dy
        r=ma.sqrt(x**2.0+y**2.0)
        phi=Phi(x,y)
        if r>0.0:
            for k in range(Nz):
                z=(k-(Nz-1)/2.0)*dz
                rho_g[k,j,i]=rho_gas(r,phi, rhop_g[k,j,i],c,Rs,h_scale)
```

```python
#-----------------------------PLOTTING
Rho=rho_g[(Nz-1)/2,:,:]

plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
plt.title("z = %0.1f au" % 0.0)
plt.pcolor(X,Y,Rho,cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/cm$^3$]')

plt.subplot(2,2,2)
plt.title("z = %0.1f au, y = %0.1f au" % (0.0, 0.0))
plt.plot(X,Rho[(N-1)/2,:])
plt.xlim(X[0],X[N-1])
plt.xlabel("X")
plt.ylabel(r'$\rho_{g}$ [g/cm$^3$]')

Rho=rho_g[:,(N-1)/2,:]

plt.subplot(2,1,2)
plt.title("y = %0.1f au" % 0.0)
plt.pcolor(X,Z,Rho,cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0],X[N-1])
plt.ylim(Z[0],Z[Nz-1])
plt.xlabel('X [au]')
plt.ylabel('Z [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/cm$^3$]')
```

## Dust distribution

For the dust distribution we used the dust density in a steady-state between mixing and drifting found in Birnstiel et al. 2013

$$ \rho(y)_{d}=C\rho_{g}\exp \left[ \frac{-\mathrm{St}(y)}{a_{\mathrm{t}}} \right]  $$
where $y=r\phi$ is the coordinate along the ring circumference, C is a normalization constant, $\mathrm{St}$ is the Stokes number which depends on $y$ via the changes in gas density and $a_{\mathrm{t}}$ as the turbulence parameter.

The Stokes number can be defined as
$$ \mathrm{St}=\frac{a\rho_{s}}{\Sigma_{g}}\frac{\pi}{2} $$
or 
$$ \mathrm{St} = \sqrt{\frac{\pi}{8}} \frac{a}{H} \frac{\rho_s}{\rho_{g}} $$

```python
C=1.0                 # Normalization factor 
alpha_t=1e-3          # dimensionless Turbulence parameter
rho_s= 1.0            # [g/cm^3]
```

```python
def rho_dust(rhogas,C,St,alpha_t):
    
    return C*rhogas*ma.exp(-St/alpha_t)
```

```python
#--------------- Dust density
N=301
Nz=51
Rmax=300              # [au]  max x or y of the sampling
Zmax=3.0*Rmax*h_scale # [au]  max z of the zampling
dx=dy=Rmax*2.0/(N-1)
dz=Zmax*2.0/(Nz-1)


a1= 1.0e-5 # [cm]
a2= 1.0e-3 # [cm]
a3= 1.0e-1 # [cm]



rho_d1=np.zeros((Nz,N,N))
rho_d2=np.zeros((Nz,N,N))
rho_d3=np.zeros((Nz,N,N))


for j in range(N):
    for i in range(N):
        x=(i-(N-1)/2.0)*dx
        y=(j-(N-1)/2.0)*dy
        r=ma.sqrt(x**2.0+y**2.0)
        phi=Phi(x,y)
        if r>0.0:
            for k in range(Nz):
                if rho_g[k,j,i]>0.0:
                    z=(k-(Nz-1)/2.0)*dz
                    St1=ma.sqrt(ma.pi/8.0)*(a1*rho_s)/(rho_g[k,j,i]*h_scale*r*au)
                    St2=ma.sqrt(ma.pi/8.0)*(a2*rho_s)/(rho_g[k,j,i]*h_scale*r*au)
                    St3=ma.sqrt(ma.pi/8.0)*(a3*rho_s)/(rho_g[k,j,i]*h_scale*r*au)

                    rho_d1[k,j,i]=rho_dust(rho_g[k,j,i],C,St1,alpha_t)
                    rho_d2[k,j,i]=rho_dust(rho_g[k,j,i],C,St2,alpha_t)
                    rho_d3[k,j,i]=rho_dust(rho_g[k,j,i],C,St3,alpha_t)
```

```python
#--------------------------PLOTTING
Rho1=rho_d1[(Nz-1)/2,:,:]
Rho2=rho_d2[(Nz-1)/2,:,:]
Rho3=rho_d3[(Nz-1)/2,:,:]



plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.subplots_adjust(wspace = 0.3)
plt.title("z = %0.1f au, a = %1.1e [cm]" % (Z[(Nz-1)/2],a1))
plt.pcolor(X,Y,Rho1, cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho1[np.nonzero(Rho1)]), vmax=np.max(Rho1[np.nonzero(Rho1)])) 
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical', label=r'$\rho_{d}$ [g/m$^3$]')

plt.subplot(1,3,2)
plt.title("z = %0.1f au, a = %1.1e [cm]" % (Z[(Nz-1)/2],a2))
plt.pcolor(X,Y,Rho2, cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho2[np.nonzero(Rho2)]), vmax=np.max(Rho2[np.nonzero(Rho2)])) 
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical', label=r'$\rho_{d}$ [g/m$^3$]')

plt.subplot(1,3,3)
plt.title("z = %0.1f au, a = %1.1e [cm]" % (Z[(Nz-1)/2],a3))
plt.pcolor(X,Y,Rho3, cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho3[np.nonzero(Rho3)]), vmax=np.max(Rho3[np.nonzero(Rho3)])) 
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical', label=r'$\rho_{d}$ [g/m$^3$]')
```
