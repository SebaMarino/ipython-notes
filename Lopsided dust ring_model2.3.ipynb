# Implementation of a Lopsided dust ring

This is an implementation of the analysis made in Birnstiel, T. et al. 2013 of a lopsided dust ring. First we construct a axi-symetric dust gas distribution
$$ \bar{\rho}_g (r,z)= \frac{\bar{\Sigma}_{g}(r)}{\sqrt{2\pi}H}\exp\left[ -\frac{z^2}{2H^2}\right] $$
where $H=c_{s}/ \Omega$ is the scale hight which is typically $0.1 r$.

$$  \bar{\Sigma}_{g}(r)=\Sigma_{c} \left(\frac{r}{r_{c}} \right) ^{-\gamma} \times \exp\left[ -\left(\frac{r}{r_{c}} \right)^{2-\gamma} \right] $$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.colors import LogNorm
import math as ma
import astropy.io.fits as pyfits
```

```python
#----------------------------------------------------PARAMETERS

au=1.496e13           # [cm]
M_sun=1.9891e33       # [g]
G=6.67259e-8          # cgs
M_star=2.5*M_sun
f_gas=0.1             # M_gas=f_gas*M_star
h_scale=0.1           # scale height H=h_scale*r
ep=0.01               # dust to gas ratio

rc=140.0
Rgap=rc*0.8
sigmac=10.0 # g/cm^2
gamma=1.0  # the more negative gamma is, the more sharp the ring is
dgap=0.01
rmin=60.0


Nx=Ny=N=51 # odd edges of the cells

Rmax=300.0             # [au]  upper edge of x or y of the sampling
Zmax=3.0*Rmax*h_scale/2.0 # [au]  upper edge of z of the sampling

dx=dy=dz=Rmax*2.0/(Nx-1)

Xedge=np.arange(-Rmax,Rmax+dx,dx) #from -Rmax to Rmax
Yedge=np.arange(-Rmax,Rmax+dx,dx) #from -Rmax to Rmax
Zedge=np.arange(-Zmax-dz/2.0,Zmax+dz/2.0+dz,dz) #from -Zmax-dz/2 to Zmax+dz/2

X=np.arange(-Rmax+dx/2.0,Rmax+dx/2.0,dx) #from -Rmax+dx/2 to Rmax-dx/2
Y=np.arange(-Rmax+dy/2.0,Rmax+dy/2.0,dx) #from -Rmax+dx/2 to Rmax-dx/2
Z=np.arange(-Zmax,Zmax+dz,dz) #from -Zmax to Zmax

Nz=len(Zedge)
```

```python
#----------------------------------------------------FUNCTIONS

def Phi(xi,yi):

    # 1st quadrant
    if xi>0.0 and yi>=0.0:
        return ma.atan(yi/xi)
    
    if xi==0.0 and yi>=0.0: return ma.pi/2.0
    
    # 2nd quadrant
    if xi<0.0 and yi>=0.0:
        return ma.pi+ma.atan(yi/xi)

    # 3rd quadrant
    if xi<0.0 and yi<0.0:
        return -ma.pi+ma.atan(yi/xi)
    
    if xi==0.0 and yi<0.0: return -ma.pi/2.0
    
    # 4th quadrant
    if xi>0.0 and yi<0.0:
        return ma.atan(yi/xi)

#-----Axisymetric density    

def Sigma_p(r,rc,sigmac,gamma):
    
    if r>rmin and r<Rgap:
        return dgap*sigmac*(r/rc)**(-gamma)*ma.exp(-((r/rc)**(2.0-gamma)))
    elif r>Rgap:
        return sigmac*(r/rc)**(-gamma)*ma.exp(-((r/rc)**(2.0-gamma))) #-1.0)/(2.0*(2.0-gamma)))
    else: return 0.0
def rho_p(r,z,h_scale,rc,sigmac,gamma):
    
    H=r*h_scale
    return Sigma_p(r,rc,sigmac,gamma)*ma.exp(-z**2.0/(2.0*H**2.0))/(ma.sqrt(2.0*ma.pi)*H*au)
```

```python
#----------------------------------------------------MAIN

rhop_g=np.zeros((Nz-1,N-1,N-1))


#--------Axisymetric solution
for j in range(N-1):
    for i in range(N-1):
        x=X[i]
        y=Y[j]
        r=ma.sqrt(x**2.0+y**2.0)
        if r>0.0:
            for k in range(Nz-1):
                z=Z[k]
            
                rhop_g[k,j,i]=rho_p(r,z,h_scale,rc,sigmac,gamma)

M_0=np.sum(rhop_g)*dx*dy*dz*(au**3.0) # g
M_gas=f_gas*M_star
rhop_g=rhop_g*M_gas/M_0 
print M_gas
```

## Non-axisymetric distribution

The non-axysimetric model is
$$  \rho_{g}(r,\phi)=\bar{\rho}_{g}(r)\left[1+A(r)\sin\left(\phi+\frac{\pi}{2}\right) \right]  $$

$$  A(r)= \frac{c-1}{c+1}\exp\left[-\frac{(r-R_{s})^2}{2H^2}\right] $$
where $c=\Sigma_{g,max}/\Sigma_{g,min}$ is the largest contrast of the gas surface density and $R_{s}$ is the position of the radial preassure bump.

```python
#----------------------------------------------------PARAMETERS

Rs=rc # [au] Position of the radial preassure bump. Rs has to be slightly higher than rc
print Rs
c=3.0                #largest contrast of the gas surface density
```

```python
#----------------------------------------------------FUNCTIONS
    
#-------Non-axisymetric density    
def A(r,c,Rs,h_scale):

    sig_bump=3.0*h_scale*Rs
    return (c-1.0)*ma.exp(-(r-Rs)**2.0/(2.0*sig_bump**2.0))/(c+1.0)
    #return (c-1.0)/(c+1.0)
def rho_gas(r,phi,rhop,c,Rs,h_scale):
    if r>rc*0.8:
        return rhop*(1.0+A(r,c,Rs,h_scale)*ma.sin(phi-ma.pi/2.0))
    else:
        return rhop
```

```python
#-------------Non-axisymetric solution

rho_g=np.zeros((Nz-1,N-1,N-1))

for j in range(N-1):
    for i in range(N-1):
        x=X[i]
        y=Y[j]
        r=ma.sqrt(x**2.0+y**2.0)
        phi=Phi(x,y)
        if r>0.0:
            for k in range(Nz-1):
                z=Z[k]
                rho_g[k,j,i]=rho_gas(r,phi, rhop_g[k,j,i],c,Rs,h_scale)

print np.sum(rho_g)*dx*dy*dz*(au**3.0) # g     
```

## Dust distribution

First we assume that the dust radial and vertical distribution is the same as for the gas.
$$ \bar{\rho}_{dust}(r,z,a)=\epsilon f(a) \bar{\rho}_{gas}(r,z) $$
where $\epsilon$ is the dust to gas ratio and $f(a)$ is the fraction of the mass in particles of size $a$.
For the dust azimuthal distribution we used the dust density in a steady-state between mixing and drifting found in Birnstiel et al. 2013

$$ \rho(\phi)_{d}=C\rho_{g}\exp \left[ \frac{-\mathrm{St}(\phi)}{a_{\mathrm{t}}} \right]  $$
where C is a normalization constant which depends on the radius and the dust abundance, $\mathrm{St}$ is the Stokes number which depends on $\phi$ via the changes in gas density and $a_{\mathrm{t}}$ as the turbulence parameter. To compute $C$ we integrate in $\phi$ with $r$ fixed imposing 
$$ \int \bar{\rho}(r)_{d} \ rd\phi = \int \rho(r,\phi)_{d} \ rd\phi = C(r,a) \int \rho(r,\phi)_{g}\exp \left[ \frac{-\mathrm{St}(\phi)}{a_{\mathrm{t}}} \right] rd\phi $$
we get
$$ C(r,a)= \frac{2\pi r \bar{\rho}(r)_{d}}{\int \rho(r,\phi)_{g}\exp \left[ \frac{-\mathrm{St}(\phi)}{a_{\mathrm{t}}} \right] rd\phi}  $$

The Stokes number can be defined as
$$ \mathrm{St}=\frac{a\rho_{s}}{\Sigma_{g}}\frac{\pi}{2} $$
or 
$$ \mathrm{St} = \sqrt{\frac{\pi}{8}} \frac{a}{H} \frac{\rho_s}{\rho_{g}} $$

```python
alpha_t=1e-2          # dimensionless Turbulence parameter
```

```python
#---------------FUNCTIONS

def rho_dust2(rhogas,C,St,Stmin,alpha_t):
    if  (St-Stmin)/alpha_t>700: return 0.0
    return C*rhogas*ma.exp(-(St-Stmin)/alpha_t)


def C_dust2(r,a,f_a,alpha_t,rhop,rho_s):
    
    
    M=500 # even
    h=2.0*ma.pi/M
    
    rg=rho_gas(r,0.0,rhop,c,Rs,h_scale)
    St_min=ma.sqrt(ma.pi/8.0)*(a*rho_s)/(rg*h_scale*r*au)
    # find max St
    for n in xrange(0,M):
        
        
        phi=2.0*ma.pi*n/M
        
        rg=rho_gas(r,phi,rhop,c,Rs,h_scale)
        St_min=min(St_min,ma.sqrt(ma.pi/8.0)*(a*rho_s)/(rg*h_scale*r*au))

    #------integral using Simpson 1/3
    I=0.0
    
    phi1=0.0
    rg1=rho_gas(r,phi1,rhop,c,Rs,h_scale)
    St_1=ma.sqrt(ma.pi/8.0)*(a*rho_s)/(rg1*h_scale*r*au)
    rd1=rho_dust2(rg1,1.0,St_1,St_min,alpha_t)
    
    for n in xrange(1,M,2):
        
        
        phi2=2.0*ma.pi*n/M
        phi3=2.0*ma.pi*(n+1)/M
        
        rg2=rho_gas(r,phi2,rhop,c,Rs,h_scale)
        rg3=rho_gas(r,phi3,rhop,c,Rs,h_scale)

        St_2=ma.sqrt(ma.pi/8.0)*(a*rho_s)/(rg2*h_scale*r*au)
        St_3=ma.sqrt(ma.pi/8.0)*(a*rho_s)/(rg3*h_scale*r*au)
        
        rd2=rho_dust2(rg2,1.0,St_2,St_min,alpha_t)
        rd3=rho_dust2(rg3,1.0,St_3,St_min,alpha_t)

        I+=h*r*( rd1 + 4.0*rd2 + rd3)/3.0
        
        rg1=rg3
        rd1=rd3
        St_1=St_3
        
    return 2.0*ma.pi*r*rhop*f_a/I,St_min
```

```python

#--------------- Dust density for different dust spiecies

# Amorphous carbon from 1.0 to 10.0 M=0.8 M_dust
# Olivine from 100.0 to 1000.0      M=0.2 M_dust
frac_ac=0.8
frac_oliv=0.2
a=np.array([1.0, 10.0, 100.0, 1000.0])*1.0e-4 # cm

rho_ac=2.0    # g/cm3
rho_oliv=4.0  # g/cm3

rho_s=[rho_ac,rho_ac,rho_oliv,rho_oliv]
f_a=np.zeros(len(a))  # fraction of the total dust mass in each species

a0_ac=a[0]
a0_oliv=a[2]

m0_ac=4.0*a0_ac**3.0/3.0*rho_ac
m0_oliv=4.0*a0_oliv**3.0/3.0*rho_oliv

for i in xrange(len(a)):
    if i<2: #-------amorphous carbon
        m=4.0*a[i]**3.0/3.0*rho_ac
        f_a[i]=ep*(a[i]/a0_ac)**(-3.5)*(m/m0_ac)

    else:  #----------Olivine
        m=4.0*a[i]**3.0/3.0*rho_oliv        
        f_a[i]=ep*(a[i]/a0_oliv)**(-3.5)*(m/m0_oliv)


f_a[2:]=f_a[2:]*(frac_oliv/frac_ac)*(np.sum(f_a[0:2]))/np.sum(f_a[2:]) # F_oliv=0.25*F_ac


f_a=f_a*ep/np.sum(f_a)
print f_a, np.sum(f_a)

rho_d=np.zeros((len(a),Nz-1,N-1,N-1))

#Calculate Cs for different rs

Cs=[]
rs=[]
St_mins=[]

for j in range(N-1):

    for i in range(N-1):
        x=(i-(N-1)/2.0)*dx+dx/2.0
        y=(j-(N-1)/2.0)*dy+dy/2.0
        r=ma.sqrt(x**2.0+y**2.0)
        H=r*h_scale
        phi=Phi(x,y)
        
        if r>0.0 and rhop_g[(Nz-2)/2,j,i]>0.0:
            
            #------ z=0
            Ctemp=np.zeros(len(a))
            St_mintemp=np.zeros(len(a))
            
            try: 
                l=rs.index(r)
                for ai in xrange(len(a)):
                    Ctemp[ai]=Cs[l][ai]
                    St_mintemp[ai]=St_mins[l][ai]
            
            except:
                rs.append(r)
                for ai in xrange(len(a)):
                    if ai<2:
                        Ctemp[ai],St_mintemp[ai]=C_dust2(r,a[ai],f_a[ai],alpha_t,rhop_g[(Nz-2)/2,j,i],rho_ac)
                    else:
                        Ctemp[ai],St_mintemp[ai]=C_dust2(r,a[ai],f_a[ai],alpha_t,rhop_g[(Nz-2)/2,j,i],rho_oliv)
                Cs.append(Ctemp)
                St_mins.append(St_mintemp)
                
            for ai in xrange(len(a)):
            
                if ai<2:
                    St=ma.sqrt(ma.pi/8.0)*(a[ai]*rho_ac)/(rho_g[(Nz-2)/2,j,i]*h_scale*r*au)
                    rho_d[ai,(Nz-2)/2,j,i]=rho_dust2(rho_g[(Nz-2)/2,j,i],Ctemp[ai],St,St_mintemp[ai],alpha_t)
                else:
                    St=ma.sqrt(ma.pi/8.0)*(a[ai]*rho_oliv)/(rho_g[(Nz-2)/2,j,i]*h_scale*r*au)
                    rho_d[ai,(Nz-2)/2,j,i]=rho_dust2(rho_g[(Nz-2)/2,j,i],Ctemp[ai],St,St_mintemp[ai],alpha_t)


            
                for k in range(0,(Nz-2)/2-1):
                
                    z=(k-Nz/2.0)*dz +dz
                
                    rho_d[ai,k,j,i]=rho_d[ai,(Nz-2)/2,j,i]*ma.exp(-z**2.0/(2.0*H*2.0))
                    rho_d[ai,Nz-2-k,j,i]=rho_d[ai,k,j,i]

# force density of dust to respect dust to gas ratio
rho_d=rho_d*ep*np.sum(rho_g)/np.sum(rho_d)
               
            
for ai in xrange(len(a)):
            
    print "M dust of %0.1f um = %1.1e M_sun" % (a[ai]*1.0e4,(np.sum(rho_d[ai,:,:,:])*(dz*dx*dy)*(au**3)/M_sun))
print "M dust = %1.4f M_gas" % (np.sum(rho_d)*(dz*dx*dy)*(au**3)/M_gas)
```
