# Implementation of a Lopsided dust ring

This is an implementation of the analysis made in Birnstiel, T. et al. 2013 of a lopsided dust ring. 

We use the following equations: 
The Stokes number can be defined as
$$ \mathrm{St}=\frac{a\rho_{s}}{\Sigma_{g}}\frac{\pi}{2} $$
or 
$$ St = \sqrt{\frac{\pi}{8}} \frac{a}{H} \frac{\rho_s}{\rho_{g}} $$

The mean density of the gas is
$$ \bar{\rho}_g (r,z)= \frac{\bar{\Sigma}_{g}(r)}{\sqrt{2\pi}H}\exp\left[ -\frac{z^2}{2H^2}\right] $$
where $H=c_{s}/ \Omega$ is the scale hight.
The mean surface density is

$$   \bar{\Sigma}_{g}(r < 90 \ au) = \delta_{gap}\Sigma_c \left(\frac{r}{r_c} \right)^{-\gamma}  $$
$$   \bar{\Sigma}_{g}(r > 90 au) = \delta_{gap}\Sigma_c \left(\frac{r}{r_c} \right)^{-\gamma}\exp\left[ \left(\frac{r}{r_c} \right)^{2-\gamma} \right]  $$

And the non-axysimetric model was
$$  \rho_{g}(r,\phi)=\bar{\rho}_{g}(r)\left[1+A(r)\sin\left(\phi+\frac{\pi}{2}\right) \right]  $$

$$  A(r)= \frac{c-1}{c+1}\exp\left[-\frac{(r-R_{s})^2}{2H^2}\right] $$
where $c=\Sigma_{g,max}/\Sigma_{g,min}$ is the largest contrast of the gas surface density and $R_{s}$ is the position of the radial preassure bump.


For the dust distribution we used the dust density in a steady-state between mixing and drifting found in Birnstiel et al. 2013

$$ \rho(y)_{d}=C\rho_{g}\exp \left[ \frac{-\mathrm{St}(y)}{a_{\mathrm{t}}} \right]  $$
where $y=r\phi$ is the coordinate along the ring circumference, C is a normalization constant, \mathrm{St} is the Stokes number which depends on $y$ via the changes in gas density and $a_{\mathrm{t}}$ as the turbulence parameter.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.colors import LogNorm
import math as ma
import astropy.io.fits as pyfits
```

```python
#----------------------------------------------------PARAMETERS

au=1.496e11 # [m]

N=301
Nz=5
Rmax=300             # [au]  max x or y of the sampling
Zmax=30              # [au] max z of the zampling
dx=dy=Rmax*2.0/(N-1)
dz=Zmax*2.0/(Nz-1)

h_scale=0.1           # scale height H=h_scale*r

Rmin=30.0            # [au] rho_gas=0 at r<Rmin
Rcavity=90.0         # [au] radius of the cavity
gap_depth=0.01        # 

Rs=100.0             # [au] Position of the radial preassure bump

c=1.5                #largest contrast of the gas surface density

gamma=1.0

r_c= 180              # [au] radius of reference for Sigmap_c
Sp_c = 1.0e5            # [g/m^2]

C=1.0                 # Normalization factor 
alpha_t=1e-2          # Turbulence parameter
```

```python
#----------------------------------------------------FUNCTIONS

def Phi(xi,yi):

    # 1st quadrant
    if xi>0.0 and yi>=0.0:
        return ma.atan(yi/xi)
    
    if xi==0.0 and yi>=0.0: return ma.pi/2.0
    
    # 2nd quadrant
    if xi<0.0 and yi>=0.0:
        return ma.pi+ma.atan(yi/xi)

    # 3rd quadrant
    if xi<0.0 and yi<0.0:
        return -ma.pi+ma.atan(yi/xi)
    
    if xi==0.0 and yi<0.0: return -ma.pi/2.0
    
    # 4th quadrant
    if xi>0.0 and yi<0.0:
        return ma.atan(yi/xi)

#-----Axisymetric density    
def Sigma_p(r,gap_depth,r_c,Sp_c,gamma):
    
    if r<Rmin:  
        return 0.0
    elif r<Rcavity:  #  inside cavity
        return gap_depth*Sp_c*(r/r_c)**(-gamma)
    else:            # outside cavity
         return Sp_c*(r/r_c)**(-gamma)*ma.exp(-(r/r_c)**(2-gamma))

def rho_p(r,z,h_scale,gap_depth,r_c,Sp_c,gamma):
    
    H=r*h_scale
    return Sigma_p(r,gap_depth,r_c,Sp_c,gamma)*ma.exp(-z**2.0/(2.0*H**2.0))/(ma.sqrt(2.0*ma.pi)*H*au)
    
    
#-------Non-axisymetric density    
def A(r,c,Rs,h_scale):
    
    H=h_scale*r
    return (c-1.0)*ma.exp(-(r-Rs)**2.0/(2.0*H**2.0))/(c+1.0)

def rho_gas(r,phi,rhop,c,Rs,h_scale):
    
    return rhop*(1.0+A(r,c,Rs,h_scale)*ma.sin(phi-ma.pi/2.0))

def rho_dust(rhogas,C,St,alpha_t):
    
    return C*rhogas*ma.exp(-St/alpha_t)
```

```python
#----------------------------------------------------MAIN

rhop_g=np.zeros((Nz,N,N))
X=np.linspace(-Rmax,Rmax,N)-dx/2.0 #esquina inferior izquierda del pixel
Y=np.linspace(-Rmax,Rmax,N)-dy/2.0
Z=np.linspace(-Zmax,Zmax,Nz)

#--------Axisymetric solution
for j in range(N):
    for i in range(N):
        x=(i-(N-1)/2.0)*dx
        y=(j-(N-1)/2.0)*dy
        r=ma.sqrt(x**2.0+y**2.0)
        if r>0.0:
            for k in range(Nz):
                z=(k-(Nz-1)/2.0)*dz
                rhop_g[k,j,i]=rho_p(r,z,h_scale,gap_depth,r_c,Sp_c,gamma)
```

```python
#-----------------------------PLOTTING
Rho=rhop_g[(Nz-1)/2,:,:]

plt.figure(figsize=(7.5,6))
plt.title("z = %0.1f au" % Z[(Nz-1)/2])
plt.pcolor(X,Y,Rho,cmap=plt.cm.YlOrRd_r, norm=LogNorm(vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)]))) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/m$^3$]')
```

```python
#-------------Non-axisymetric solution

rho_g=np.zeros((Nz,N,N))


for j in range(N):
    for i in range(N):
        x=(i-(N-1)/2.0)*dx
        y=(j-(N-1)/2.0)*dy
        r=ma.sqrt(x**2.0+y**2.0)
        phi=Phi(x,y)
        if r>0.0:
            for k in range(Nz):
                z=(k-(Nz-1)/2.0)*dz
                rho_g[k,j,i]=rho_gas(r,phi, rhop_g[k,j,i],c,Rs,h_scale)
```

```python
#-----------------------------PLOTTING
Rho=rho_g[(Nz-1)/2,:,:]

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.title("z = %0.1f au" % Z[(Nz-1)/2])
plt.pcolor(X,Y,Rho,cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/m$^3$]')

plt.subplot(1,2,2)
plt.title("z = %0.1f au, y = %0.1f au" % (Z[(Nz-1)/2], Y[(Nz-1)/2]))
plt.plot(X,Rho[(N-1)/2,:])
plt.xlim(X[0],X[N-1])
plt.xlabel("X")
plt.ylabel(r'$\rho_{g}$ [g/m$^3$]')
```

```python
#--------------- Dust density

a1= 1.0e-6 # [m]
a2= 1.0e-2 # [m]
a3= 1.0e-1 # [m]

rho_s= 0.5e6  # [g/m^3]

rho_d1=np.zeros((Nz,N,N))
rho_d2=np.zeros((Nz,N,N))
rho_d3=np.zeros((Nz,N,N))


for j in range(N):
    for i in range(N):
        x=(i-(N-1)/2.0)*dx
        y=(j-(N-1)/2.0)*dy
        r=ma.sqrt(x**2.0+y**2.0)
        phi=Phi(x,y)
        if r>0.0:
            for k in range(Nz):
                if rho_g[k,j,i]>0.0:
                    z=(k-(Nz-1)/2.0)*dz
                    St1=ma.sqrt(ma.pi/8.0)*(a1*rho_s)/(rho_g[k,j,i]*h_scale*r*au)
                    St2=ma.sqrt(ma.pi/8.0)*(a2*rho_s)/(rho_g[k,j,i]*h_scale*r*au)
                    St3=ma.sqrt(ma.pi/8.0)*(a3*rho_s)/(rho_g[k,j,i]*h_scale*r*au)

                    rho_d1[k,j,i]=rho_dust(rho_g[k,j,i],C,St1,alpha_t)
                    rho_d2[k,j,i]=rho_dust(rho_g[k,j,i],C,St2,alpha_t)
                    rho_d3[k,j,i]=rho_dust(rho_g[k,j,i],C,St3,alpha_t)
```

```python
#--------------------------PLOTTING
Rho1=rho_d1[(Nz-1)/2,:,:]
Rho2=rho_d2[(Nz-1)/2,:,:]
Rho3=rho_d3[(Nz-1)/2,:,:]



plt.figure(figsize=(18,4))
plt.subplot(1,3,1)
plt.subplots_adjust(wspace = 0.3)
plt.title("z = %0.1f au, a = %1.1e [m]" % (Z[(Nz-1)/2],a1))
plt.pcolor(X,Y,Rho1, cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho1[np.nonzero(Rho1)]), vmax=np.max(Rho1[np.nonzero(Rho1)])) 
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical', label=r'$\rho_{d}$ [g/m$^3$]')

plt.subplot(1,3,2)
plt.title("z = %0.1f au, a = %1.1e [m]" % (Z[(Nz-1)/2],a2))
plt.pcolor(X,Y,Rho2, cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho2[np.nonzero(Rho2)]), vmax=np.max(Rho2[np.nonzero(Rho2)])) 
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical', label=r'$\rho_{d}$ [g/m$^3$]')

plt.subplot(1,3,3)
plt.title("z = %0.1f au, a = %1.1e [m]" % (Z[(Nz-1)/2],a3))
plt.pcolor(X,Y,Rho3, cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho3[np.nonzero(Rho3)]), vmax=np.max(Rho[np.nonzero(Rho3)])) 
plt.xlim(X[0],X[N-1])
plt.ylim(Y[0],Y[N-1])
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical', label=r'$\rho_{d}$ [g/m$^3$]')
```

```python
# Contrast at the same orbit

#Rs=50.0
r=Rs
rho_s= 0.5e6  # [g/m^3]

phis=np.linspace(0,2.0*ma.pi,1000)

rg=np.zeros(1000)

rd1=np.zeros(1000)
rd2=np.zeros(1000)
rd3=np.zeros(1000)

a1=1e-6
a2=1e-3
a3=1e-1

for i in range(1000):
    rg[i]=rho_gas(r,phis[i],1.0e-8,c,Rs,h_scale)
    St1=ma.sqrt(ma.pi/8.0)*(a1*rho_s)/(rg[i]*h_scale*r*au)
    St2=ma.sqrt(ma.pi/8.0)*(a2*rho_s)/(rg[i]*h_scale*r*au)
    St3=ma.sqrt(ma.pi/8.0)*(a3*rho_s)/(rg[i]*h_scale*r*au)

    rd1[i]=rho_dust(rg[i],1.0,St1,alpha_t)
    rd2[i]=rho_dust(rg[i],1.0,St2,alpha_t)
    rd3[i]=rho_dust(rg[i],1.0,St3,alpha_t)
    
rg=rg/rg[500]
rd1=rd1/rd1[500]
rd2=rd2/rd2[500]
rd3=rd3/rd3[500]
```

```python
#--------------------------PLOTTING

plt.figure(figsize=(8,6))
plt.plot(phis,rg,color="red",label='gas')
plt.plot(phis,rd1,color="blue", label='dust a = %0.1e m' % a1)
plt.plot(phis,rd2,color="green", label='dust a = %0.1e m' % a2)
plt.plot(phis,rd3,color="black", label='dust a = %0.1e m' % a3)
plt.xlim(phis[0],phis[999])
plt.ylim(1e-2,1e1)
plt.yscale('log')
plt.legend(loc=1)
#plt.ylim(0.0,1.5)
```
