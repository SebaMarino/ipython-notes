# Implementation of a Lopsided dust ring

This is an implementation of the analysis made in Birnstiel, T. et al. 2013 of a lopsided dust ring. First we construct a axi-symetric dust gas distribution
$$ \bar{\rho}_g (r,z)= \frac{\bar{\Sigma}_{g}(r)}{\sqrt{2\pi}H}\exp\left[ -\frac{z^2}{2H^2}\right] $$
where $H=c_{s}/ \Omega$ is the scale hight which is typically $0.1 r$.

$$ \begin{align}  \bar{\Sigma}_{g}(r < 90 \ au) &=& \delta_{gap}\Sigma_c \left(\frac{r}{r_c} \right)^{-\gamma}  \\
   \bar{\Sigma}_{g}(r > 90 au) &=& \Sigma_c \left(\frac{r}{r_c} \right)^{-\gamma}\exp\left[ \left(\frac{r}{r_c} \right)^{2-\gamma} \right]  \end{align}$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.colors import LogNorm
import math as ma
import astropy.io.fits as pyfits
```

```python
#----------------------------------------------------PARAMETERS

au=1.496e13           # [cm]

h_scale=0.1           # scale height H=h_scale*r
Rmin=30.0             # [au] rho_gas=0 at r<Rmin
Rcavity=90.0          # [au] radius of the cavity
gap_depth=0.01        # decrement in the surface density inside the cavity
gamma=1.0
r_c= 180              # [au] radius of reference for Sigmap_c
Sp_c = 10.0           # [g/cm^2]

N=101
Nz=21
Rmax=300              # [au]  max x or y of the sampling
Zmax=3.0*Rmax*h_scale # [au]  max z of the zampling
dx=dy=Rmax*2.0/(N-1)
dz=Zmax*2.0/(Nz-1)
```

```python
#----------------------------------------------------FUNCTIONS

def Phi(xi,yi):

    # 1st quadrant
    if xi>0.0 and yi>=0.0:
        return ma.atan(yi/xi)
    
    if xi==0.0 and yi>=0.0: return ma.pi/2.0
    
    # 2nd quadrant
    if xi<0.0 and yi>=0.0:
        return ma.pi+ma.atan(yi/xi)

    # 3rd quadrant
    if xi<0.0 and yi<0.0:
        return -ma.pi+ma.atan(yi/xi)
    
    if xi==0.0 and yi<0.0: return -ma.pi/2.0
    
    # 4th quadrant
    if xi>0.0 and yi<0.0:
        return ma.atan(yi/xi)

#-----Axisymetric density    
def Sigma_p(r,gap_depth,r_c,Sp_c,gamma):
    
    if r<Rmin:  
        return 0.0
    elif r<Rcavity:  #  inside cavity
        return gap_depth*Sp_c*(r/r_c)**(-gamma)
    else:            # outside cavity
         return Sp_c*(r/r_c)**(-gamma)*ma.exp(-(r/r_c)**(2-gamma))

def rho_p(r,z,h_scale,gap_depth,r_c,Sp_c,gamma):
    
    H=r*h_scale
    return Sigma_p(r,gap_depth,r_c,Sp_c,gamma)*ma.exp(-z**2.0/(2.0*H**2.0))/(ma.sqrt(2.0*ma.pi)*H*au)
    
```

```python
#----------------------------------------------------MAIN

rhop_g=np.zeros((Nz,N,N))
X=np.linspace(-Rmax,Rmax+dx,N+1) 
Y=np.linspace(-Rmax,Rmax+dy,N+1)
Z=np.linspace(-Zmax,Zmax+dz,Nz+1)

#--------Axisymetric solution
for j in range(N):
    for i in range(N):
        x=(i-(N-1)/2.0)*dx
        y=(j-(N-1)/2.0)*dy
        r=ma.sqrt(x**2.0+y**2.0)
        if r>0.0:
            for k in range(Nz):
                z=(k-(Nz-1)/2.0)*dz
            
                rhop_g[k,j,i]=rho_p(r,z,h_scale,gap_depth,r_c,Sp_c,gamma)
```

```python

#-----------------------------PLOTTING
Rho=rhop_g[(Nz-1)/2,:,:]

plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
plt.title("z = %0.1f au" % 0.0)
plt.pcolor(X-dx/2.0,Y-dy/2.0,Rho,cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0]-dx/2.0,X[N]-dx/2.0)
plt.ylim(Y[0]-dy/2.0,Y[N]-dy/2.0)
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/cm$^3$]')

plt.subplot(2,2,2)
plt.title("z = %0.1f au, y = %0.1f au" % (0.0, 0.0))
plt.plot(X[:-1],Rho[(N-1)/2,:])
plt.xlim(X[0],X[N-1])
plt.xlabel("X")
plt.ylabel(r'$\rho_{g}$ [g/cm$^3$]')

Rho=rhop_g[:,(N-1)/2,:]

plt.subplot(2,1,2)
plt.title("y = %0.1f au" % 0.0)
plt.pcolor(X-dx/2.0,Z-dz/2.0,Rho,cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0]-dx/2.0,X[N]-dx/2.0)
plt.ylim(Z[0]-dz/2.0,Z[Nz]-dz/2.0)
plt.xlabel('X [au]')
plt.ylabel('Z [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/cm$^3$]')
```

## Non-axisymetric distribution

The non-axysimetric model is
$$  \rho_{g}(r,\phi)=\bar{\rho}_{g}(r)\left[1+A(r)\sin\left(\phi+\frac{\pi}{2}\right) \right]  $$

$$  A(r)= \frac{c-1}{c+1}\exp\left[-\frac{(r-R_{s})^2}{2H^2}\right] $$
where $c=\Sigma_{g,max}/\Sigma_{g,min}$ is the largest contrast of the gas surface density and $R_{s}$ is the position of the radial preassure bump.

```python
#----------------------------------------------------PARAMETERS

Rs=120.0             # [au] Position of the radial preassure bump
c=5.0                #largest contrast of the gas surface density
```

```python
#----------------------------------------------------FUNCTIONS
    
#-------Non-axisymetric density    
def A(r,c,Rs,h_scale):
    
    H=h_scale*r
    #return (c-1.0)*ma.exp(-(r-Rs)**2.0/(2.0*H**2.0))/(c+1.0)
    return (c-1.0)/(c+1.0)
def rho_gas(r,phi,rhop,c,Rs,h_scale):
    
    return rhop*(1.0+A(r,c,Rs,h_scale)*ma.sin(phi-ma.pi/2.0))
```

```python
#-------------Non-axisymetric solution

rho_g=np.zeros((Nz,N,N))


for j in range(N):
    for i in range(N):
        x=(i-(N-1)/2.0)*dx
        y=(j-(N-1)/2.0)*dy
        r=ma.sqrt(x**2.0+y**2.0)
        phi=Phi(x,y)
        if r>0.0:
            for k in range(Nz):
                z=(k-(Nz-1)/2.0)*dz
                rho_g[k,j,i]=rho_gas(r,phi, rhop_g[k,j,i],c,Rs,h_scale)
```

```python
#-----------------------------PLOTTING
Rho=rho_g[(Nz-1)/2,:,:]

plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
plt.title("z = %0.1f au" % 0.0)
plt.pcolor(X-dx/2.0,Y-dy/2.0,Rho,cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0]-dx/2.0,X[N]-dx/2.0)
plt.ylim(Y[0]-dy/2.0,Y[N]-dy/2.0)
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/cm$^3$]')

plt.subplot(2,2,2)
plt.title("z = %0.1f au, y = %0.1f au" % (0.0, 0.0))
plt.plot(X[:-1],Rho[(N-1)/2,:])
plt.xlim(X[0],X[N-1])
plt.xlabel("X")
plt.ylabel(r'$\rho_{g}$ [g/cm$^3$]')

Rho=rho_g[:,(N-1)/2,:]

plt.subplot(2,1,2)
plt.title("y = %0.1f au" % 0.0)
plt.pcolor(X-dx/2.0,Z-dz/2.0,Rho,cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho[np.nonzero(Rho)]), vmax=np.max(Rho[np.nonzero(Rho)])) #norm=LogNorm(vmin=1e-15, vmax=np.max(rho_disk) ) )
plt.xlim(X[0]-dx/2.0,X[N]-dx/2.0)
plt.ylim(Z[0]-dz/2.0,Z[Nz]-dz/2.0)
plt.xlabel('X [au]')
plt.ylabel('Z [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='vertical',label=r'$\rho_{g}$ [g/cm$^3$]')
```

## Dust distribution

First we assume that the dust radial and vertical distribution is the same as for the gas.
$$ \bar{\rho}_{dust}(r,z,a)=\epsilon f(a) \bar{\rho}_{gas}(r,z) $$
where $\epsilon$ is the dust to gas ratio and $f(a)$ is the fraction of the mass in particles of size $a$.
For the dust azimuthal distribution we used the dust density in a steady-state between mixing and drifting found in Birnstiel et al. 2013

$$ \rho(\phi)_{d}=C\rho_{g}\exp \left[ \frac{-\mathrm{St}(\phi)}{a_{\mathrm{t}}} \right]  $$
where C is a normalization constant which depends on the radius and the dust abundance, $\mathrm{St}$ is the Stokes number which depends on $y$ via the changes in gas density and $a_{\mathrm{t}}$ as the turbulence parameter. To compute $C$ we integrate in $\phi$ with $r$ fixed imposing 
$$ \int \bar{\rho}(r)_{d} \ rd\phi = \int \rho(r,\phi)_{d} \ rd\phi = C(r,a) \int \rho(r,\phi)_{g}\exp \left[ \frac{-\mathrm{St}(\phi)}{a_{\mathrm{t}}} \right] rd\phi $$
we get
$$ C(r,a)= \frac{2\pi r \bar{\rho}(r)_{d}}{\int \rho(r,\phi)_{g}\exp \left[ \frac{-\mathrm{St}(\phi)}{a_{\mathrm{t}}} \right] rd\phi}  $$

The Stokes number can be defined as
$$ \mathrm{St}=\frac{a\rho_{s}}{\Sigma_{g}}\frac{\pi}{2} $$
or 
$$ \mathrm{St} = \sqrt{\frac{\pi}{8}} \frac{a}{H} \frac{\rho_s}{\rho_{g}} $$

```python
alpha_t=1e-3          # dimensionless Turbulence parameter
rho_s= 1.0            # [g/cm^3]
ep=0.01               # dust to gas ratio
```

```python
def rho_dust(rhogas,C,St,alpha_t):
    
    return C*rhogas*ma.exp(-St/alpha_t)

def C_dust(r,a,ep,f_a,alpha_t,rhop):
    
    
    M=10000 # even
    h=ma.pi/M

    #------integral ussing Simpson 1/3
    I=0.0
    
    phi1=0.0
    rg1=rho_gas(r,phi1,rhop,c,Rs,h_scale)
    St_1=ma.sqrt(ma.pi/8.0)*(a*rho_s)/(rg1*h_scale*r*au)
    rd1=rho_dust(rg1,1.0,St_1,alpha_t)
    
    for n in xrange(1,M,2):
        
        
        phi2=ma.pi*n/M
        phi3=ma.pi*(n+1)/M
        
        rg2=rho_gas(r,phi2,rhop,c,Rs,h_scale)
        rg3=rho_gas(r,phi3,rhop,c,Rs,h_scale)

        St_2=ma.sqrt(ma.pi/8.0)*(a*rho_s)/(rg2*h_scale*r*au)
        St_3=ma.sqrt(ma.pi/8.0)*(a*rho_s)/(rg3*h_scale*r*au)
        
        rd2=rho_dust(rg1,1.0,St_2,alpha_t)
        rd3=rho_dust(rg1,1.0,St_3,alpha_t)

        I+=h*r*(rd1+4.0*rd2+rd3)/3.0
        
        rd1=rd3
    
    return 2.0*ma.pi*r*rhop*ep*f_a/I
        
```

```python
#--------------- Dust density for three different dust spiecies



a1= 1.0e-7 # [cm]
a2= 1.0e-4 # [cm]
a3= 1.0e-1 # [cm]

f_a1=1.0
f_a2=f_a1*(a2/a1)**(-3.5)
f_a3=f_a1*(a3/a1)**(-3.5)


rho_d1=np.zeros((Nz,N,N))
rho_d2=np.zeros((Nz,N,N))
rho_d3=np.zeros((Nz,N,N))



#Calculate Cs for different rs
Cs=[]
rs=[]

for j in range(N):
    for i in range(N):
        x=(i-(N-1)/2.0)*dx
        y=(j-(N-1)/2.0)*dy
        r=ma.sqrt(x**2.0+y**2.0)
        H=r*h_scale
        phi=Phi(x,y)
        if r>0.0 and rhop_g[(Nz-1)/2,j,i]>10**-15:
            
            #------ z=0
            
            
            try: 
                l=rs.index(r)
                C1=Cs[l][0]
                C2=Cs[l][1]
                C3=Cs[l][2]
                
            
            except:
                """
                l=min(range(len(rs)), key=lambda n: abs(rs[n]-r))
                C1=Cs[l,0]
                C2=Cs[l,1]
                C3=Cs[l,2]
                
                if rs[l]>r:
                    C1=(Cs[l,0]-Cs[l-1,0])*(r-rs[l-1])/(dx)+Cs[l-1,0]
                    C2=(Cs[l,1]-Cs[l-1,1])*(r-rs[l-1])/(dx)+Cs[l-1,1]
                    C3=(Cs[l,2]-Cs[l-1,2])*(r-rs[l-1])/(dx)+Cs[l-1,2]

                else: 
                    C1=(Cs[l+1,0]-Cs[l,0])*(r-rs[l])/(dx)+Cs[l,0]
                    C2=(Cs[l+1,1]-Cs[l,1])*(r-rs[l])/(dx)+Cs[l,1]
                    C3=(Cs[l+1,2]-Cs[l,2])*(r-rs[l])/(dx)+Cs[l,2]                   
                """
            
            
                C1=C_dust(r,a1,ep,f_a1,alpha_t,rhop_g[(Nz-1)/2,j,i])
                C2=C_dust(r,a2,ep,f_a2,alpha_t,rhop_g[(Nz-1)/2,j,i])
                C3=C_dust(r,a3,ep,f_a3,alpha_t,rhop_g[(Nz-1)/2,j,i])
                rs.append(r)
                Cs.append([C1,C2,C3])
              
            St1=ma.sqrt(ma.pi/8.0)*(a1*rho_s)/(rho_g[(Nz-1)/2,j,i]*h_scale*r*au)
            St2=ma.sqrt(ma.pi/8.0)*(a2*rho_s)/(rho_g[(Nz-1)/2,j,i]*h_scale*r*au)
            St3=ma.sqrt(ma.pi/8.0)*(a3*rho_s)/(rho_g[(Nz-1)/2,j,i]*h_scale*r*au)

            rho_d1[(Nz-1)/2,j,i]=rho_dust(rho_g[(Nz-1)/2,j,i],C1,St1,alpha_t)
            rho_d2[(Nz-1)/2,j,i]=rho_dust(rho_g[(Nz-1)/2,j,i],C2,St2,alpha_t)
            rho_d3[(Nz-1)/2,j,i]=rho_dust(rho_g[(Nz-1)/2,j,i],C3,St3,alpha_t)
            
            #------ z!=0
            
            for k in range(0,(Nz-1)/2):
                
                z=(k-(Nz-1)/2.0)*dz
                
                rho_d1[k,j,i]=rho_d1[(Nz-1)/2,j,i]*ma.exp(-z**2.0/(2.0*H*2.0))
                rho_d1[Nz-1-k,j,i]=rho_d1[k,j,i]

                rho_d2[k,j,i]=rho_d2[(Nz-1)/2,j,i]*ma.exp(-z**2.0/(2.0*H*2.0))
                rho_d2[Nz-1-k,j,i]=rho_d2[k,j,i]
                
                rho_d3[k,j,i]=rho_d3[(Nz-1)/2,j,i]*ma.exp(-z**2.0/(2.0*H*2.0))
                rho_d3[Nz-1-k,j,i]=rho_d3[k,j,i]                   
            
```

```python
#--------------------------PLOTTING
Rho1=rho_d1[(Nz-1)/2,:,:]
Rho2=rho_d2[(Nz-1)/2,:,:]
Rho3=rho_d3[(Nz-1)/2,:,:]


plt.figure(figsize=(18,8))
plt.subplot(1,3,1)
plt.subplots_adjust(wspace = 0.3)
plt.title("z = %0.1f au, a = %1.1e [cm]" % (Z[(Nz-1)/2],a1))
plt.pcolor(X-dx/2.0,Y-dy/2.0,Rho1, cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho1[np.nonzero(Rho1)]), vmax=np.max(Rho1[np.nonzero(Rho1)])) 
plt.xlim(X[0]-dx/2.0,X[N]-dx/2.0)
plt.ylim(Y[0]-dy/2.0,Y[N]-dy/2.0)
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='horizontal', label=r'$\rho_{d}$ [g/cm$^3$]')

plt.subplot(1,3,2)
plt.title("z = %0.1f au, a = %1.1e [cm]" % (Z[(Nz-1)/2],a2))
plt.pcolor(X-dx/2.0,Y-dy/2.0,Rho2, cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho2[np.nonzero(Rho2)]), vmax=np.max(Rho2[np.nonzero(Rho2)])) 
plt.xlim(X[0]-dx/2.0,X[N]-dx/2.0)
plt.ylim(Y[0]-dy/2.0,Y[N]-dy/2.0)
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='horizontal', label=r'$\rho_{d}$ [g/cm$^3$]')

plt.subplot(1,3,3)
plt.title("z = %0.1f au, a = %1.1e [cm]" % (Z[(Nz-1)/2],a3))
plt.pcolor(X-dx/2.0,Y-dy/2.0,Rho3, cmap=plt.cm.YlOrRd_r, vmin=np.min(Rho3[np.nonzero(Rho3)]), vmax=np.max(Rho3[np.nonzero(Rho3)])) 
plt.xlim(X[0]-dx/2.0,X[N]-dx/2.0)
plt.ylim(Y[0]-dy/2.0,Y[N]-dy/2.0)
plt.xlabel('X [au]')
plt.ylabel('Y [au]')
fig = plt.gcf()
circle=plt.Circle((0,0),Rmin,color="black",fill=True)
fig.gca().add_artist(circle)
plt.colorbar(orientation='horizontal', label=r'$\rho_{d}$ [g/cm$^3$]')
```
